## Importando a biblioteca

```python
import matplotlib.pyplot as plt
```

## Primeiro gráfico

```python
# As coisas aqui são tão simples de entender que eu não vou nem explicar tanto

x = [1, 2, 3, 4, 5]
y = [2, 5, 9, 3, 8]


plt.plot(x, y, label="Relação entre X e Y", color='red', linestyle='--', marker='o')

plt.title("Gráfico de linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()

plt.show()
```

### E se eu quiser plotar mais de uma linha?

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 9, 3, 8]
y2 = [1, 2, 1, 1, 2]


plt.plot(x, y1, label="Relação entre X e Y1", color='red', linestyle='--', marker='o')
plt.plot(x, y2, label="Relação entre X e Y2", color='blue', linestyle='-', marker='x')

plt.title("Gráfico de linha")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()

plt.show()
```

## Gráfico de dispersão

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 5, 9, 3, 8]
y2 = [1, 2, 1, 1, 2]

# "s=100" muda o tamanho do marcador no gráfico
plt.scatter(x, y1, label="Relação entre X e Y1", color='red', linestyle='--', marker='o', s=50)
plt.scatter(x, y2, label="Relação entre X e Y2", color='blue', linestyle='-', marker='x')

# isso deve servir pra algo
plt.fill_between(x, y1, y2, color='orange')

plt.title("Gráfico de disperssão")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()

plt.show()
```

## Gráfico de barras

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]

# Se quiser o gráfico na horizontal, use barh()
plt.bar(categorias, valores)
plt.show()
```

## Gráfico de pizza

```python
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]

# Esse autopct="%1.1f%%" serve para por a porcentagem no gráfico de pizza
plt.pie(valores, labels=categorias, autopct="%1.1f%%")
plt.show()
```

## Importando a biblioteca pandas

## Lendo os dados

```python
# vamos ler esses dados e organizar eles para serem plotados
dados = pd.read_csv('dados.csv')
dados
```

```python
# separando a coluna Date
dias = dados['Date']
dias
```

```python
# separando a coluna PETR4
petro = dados['PETR4']
petro
```

```python
# separando a coluna ITUB4
itau = dados['ITUB4']
itau
```

```python
# agora é só plotar as informações
plt.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')
plt.plot(dias, itau, label='Itau', color='orange', linestyle='--', marker='.')

plt.title('Cotação -2025/1')
plt.xlabel('Dias')
plt.ylabel('Valor da ação (R$)')
plt.legend()

plt.show()

# só que dessa forma você vai ver que ficou bem bagunçado
```

```python
# então é melhor fazer assim 

# só com essa linha ja melhora muito, mas ainda temos o problema da data sobreposta
fig = plt.figure(figsize=(12, 8))

plt.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')
plt.plot(dias, itau, label='Itau', color='orange', linestyle='--', marker='.')

plt.title('Cotação - 2025/1')
plt.xlabel('Dias')
plt.ylabel('Valor da ação (R$)')
plt.grid(True)
plt.legend()

# então com essa linha aqui conseguimos arrumar a data (O ::20 é para pular de 20 em 20 dias)
plt.xticks(ticks=dias[::20])

plt.show()
```

```python
# com o pandas também podemos analisar o aumento e o declínio dos valores
# uma forma boa de ver isso é usando histograma
petro.diff()
```

## Histograma

```python
fig = plt.figure(figsize=(12, 8))

plt.hist(petro.diff(), label='Petro', color='green', alpha=0.5, bins=50)
plt.axvline(petro.diff().mean(), color='green', linestyle='--', linewidth=4)

plt.hist(itau.diff(), label='Itau', color='orange', alpha=0.5, bins=50)
plt.axvline(itau.diff().mean(), color='orange', linestyle='--', linewidth=4)

plt.title('Variação diária - 2025/1')
plt.xlabel('Variação (R$)')
plt.ylabel('Frequência')
plt.legend()

plt.show()
```

### E se eu quiser plotar dois gráficos numa mesma figura?

```python
# essa linha serve para usar um estilo pronto que a propria biblioteca fornece
plt.style.use('dark_background')

# plotando gráficos vazios
fig, (ax_cima, ax_baixo) = plt.subplots(figsize=(12, 8), nrows=2, sharex=True)

ax_cima.hist(petro.diff(), label='Petro', color='green', bins=50)
ax_cima.axvline(petro.diff().mean(), color='gray', linestyle='--', linewidth=4)
ax_cima.set_xlabel('Variação (R$)')
ax_cima.set_ylabel('Frequência')
ax_cima.grid(True)

ax_baixo.hist(itau.diff(), label='Itau', color='orange', bins=50)
ax_baixo.axvline(itau.diff().mean(), color='gray', linestyle='--', linewidth=4)
ax_baixo.set_xlabel('Variação (R$)')
ax_baixo.set_ylabel('Frequência')
ax_baixo.grid(True)

fig.suptitle('Variação diária - 2025/1')
fig.legend()

plt.show()
```

## Plotagem final

```python
# cria uma figura do matplotlib, com tamanho 12x8 polegadas
fig = plt.figure(figsize=(12, 8))

# cria uma grade (grid) com 2 linhas e 2 colunas.
# essa grade permite posicionar subplots exatamente no lugar desejado
grid = fig.add_gridspec(2, 2)

# cria o primeiro subplot ocupando toda a primeira linha da grade (linha 0, todas as colunas)
ax_linha = fig.add_subplot(grid[0, :])

# cria o segundo subplot na linha 1, coluna 0
ax_hist1 = fig.add_subplot(grid[1, 0])

# cria o terceiro subplot na linha 1, coluna 1
ax_hist2 = fig.add_subplot(grid[1, 1])


# ======= GRÁFICO DE LINHA =======

# plota a série de valores da Petro
# marker='.' coloca um pontinho em cada valor
# linestyle='--' coloca linhas tracejadas
# color='green' deixa a linha verde
ax_linha.plot(dias, petro, label='Petro', color='green', linestyle='--', marker='.')

# plota a série do Itaú no mesmo gráfico
ax_linha.plot(dias, itau, label='Itau', color='orange', linestyle='--', marker='.')

# título do gráfico principal
ax_linha.set_title('Cotação - 2025/1')

# nomes dos eixos
ax_linha.set_xlabel('Dias')
ax_linha.set_ylabel('Valor da ação (R$)')

# ativa a grade de fundo
ax_linha.grid(True)

# define os valores mostrados no eixo X a cada 20 dias
ax_linha.set_xticks(ticks=dias[::20])


# ======= HISTOGRAMA 1 (variação Petro) =======

# histograma das variações da Petro (diferença entre os dias)
ax_hist1.hist(petro.diff(), label='Petro', color='green', bins=50)

# linha vertical indicando a média da variação
ax_hist1.axvline(petro.diff().mean(), color='gray', linestyle='--', linewidth=4)

# nomes dos eixos
ax_hist1.set_xlabel('Variação (R$)')
ax_hist1.set_ylabel('Frequência')

# grade de fundo
ax_hist1.grid(True)


# ======= HISTOGRAMA 2 (variação Itaú) =======

# histograma das variações do Itaú
ax_hist2.hist(itau.diff(), label='Itau', color='orange', bins=50)

# linha vertical da média da variação
ax_hist2.axvline(itau.diff().mean(), color='gray', linestyle='--', linewidth=4)

# nomes dos eixos
ax_hist2.set_xlabel('Variação (R$)')
ax_hist2.set_ylabel('Frequência')

# grade de fundo
ax_hist2.grid(True)


# ======= TÍTULO GERAL E AJUSTES =======

# título geral da figura
fig.suptitle('Análise de ações')

# legenda para os gráficos (usa os labels definidos nos plots)
fig.legend()

# ajusta automaticamente espaçamentos para evitar sobreposição
fig.tight_layout()

plt.show()
```
